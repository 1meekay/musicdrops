```python
# grab artist names
# (?<='name': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=', 'ty)

# grab artist id (use for uri and url)
# (?<='id': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=', 'n)

# ===========
# grab artist url (can use baseURL + artist id) --> https://open.spotify.com/artist/{artist id}
# (?<='spotify': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=')
# ===========

# grab album name
# (?<='name': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=', 'release_date')

# grab album id (use for uri and url)
# (?<='id': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=', 'i)

# ===========
# grab album url (can use baseURL + album id) --> https://open.spotify.com/album/{album id}
# (?<='spotify': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*album[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=')
# ===========

# grab album art (large)
# (?<='url': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=', 'width': 640)

# grab release date
# (?<='release_date': ')[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=', 'r)

# grab total tracks (test this against len())
# (?<='total_tracks': )[\w\s!\"#$%&\'()*+,\-.\/:;<=>?@\[\]\\_\^`\|~]*(?=, 't)

# ===========
# artist uri: spotify:artist:artist id
# album uri: spotify:album:album id
```
